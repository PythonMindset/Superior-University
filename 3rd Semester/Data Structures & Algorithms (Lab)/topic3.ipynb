```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None

class Linked_list:
    def __init__(self):
        self.head=None

    def add_start(self,data):
        new_node=Node(data)
        new_node.ref=self.head
        self.head=new_node

    def add_last(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.ref!=None:
                n=n.ref
            n.ref=new_node

    def insert_at_specific_position(self,data,position):
        if position<1:
            print("Invalid position!")
            return
        new_node=Node(data)
        if position==1:
            new_node.ref=self.head
            self.head=new_node
            return
        n=self.head
        for i in range(position-2):
            if n is None:
                print("Position out of bounds")
                return
            n=n.ref
        new_node.ref=n.ref
        n.ref=new_node
    
    def delete_by_value(self,value):
        if self.head is None:
            print("List is empty")
            return
        if self.head.data==value:
            self.head=self.head.ref
            return
        n=self.head
        while n.ref is not None:
            if n.ref.data==value:
                n.ref=n.ref.ref
                return
            n=n.ref
        print("Value not found in the list")
    
    def search_by_value(self,value):
        n=self.head
        position=1
        while n is not None:
            if n.data==value:
                return position
            n=n.ref
            position+=1
        return -1
    
    def display(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            n=self.head
            while n!=None:
                print(n.data)
                n=n.ref

obj1=Linked_list()
obj1.add_start(10)
obj1.add_start(20)
obj1.add_start(30)
obj1.add_last(40)
obj1.insert_at_specific_position(50,2)
obj1.delete_by_value(40)
obj1.search_by_value(20)
obj1.display()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None

class Linked_list:
    def __init__(self):
        self.head=None

    def add_last(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.ref is not None:
                temp=temp.ref
            temp.ref=new_node
        return new_node
    
    def create_loop(self, position):
        if position<1:
            return
        loop_start=None
        temp=self.head
        count=1
        while temp.ref is not None:
            if count==position:
                loop_start = temp
            temp=temp.ref
            count+=1
        if loop_start:
            temp.ref=loop_start
    
    def detect_loop(self):
        slow = self.head
        fast = self.head
        while fast is not None and fast.ref is not None:
            slow=slow.ref
            fast=fast.ref.ref
            if slow==fast:
                return True
        return False
    
    def find_loop_start(self):
        slow=self.head
        fast=self.head

        while fast and fast.ref:
            slow=slow.ref
            fast=fast.ref.ref

            if slow==fast:
                slow=self.head
                while slow!=fast:
                    slow=slow.ref
                    fast=fast.ref
                return slow
        return None
    
    def remove_loop(self):
        loop_start=self.find_loop_start()
        if loop_start is None:
            return
        temp=loop_start
        while temp.ref!=loop_start:
            temp=temp.ref
        temp.ref=None
    
    def display(self, limit=15):
        visited=set()
        temp=self.head
        count=0
        while temp and count<limit:
            if temp in visited:
                print(f"{temp.data} (loop starts here)")
                break
            print(temp.data,end=" -> ")
            visited.add(temp)
            temp=temp.ref
            count+=1
        else:
            print("None")
    
obj1 = Linked_list()
obj1.add_last(10)
obj1.add_last(20)
node30 = obj1.add_last(30)
obj1.add_last(40)
obj1.add_last(50)

obj1.create_loop(3)

print("--> Linked List Before Removing Loop:")
obj1.display()

print("\nLoop detected:", obj1.detect_loop())

loop_node = obj1.find_loop_start()
if loop_node:
    print("Loop starts at node:", loop_node.data)

obj1.remove_loop()

print("\nLoop detected after removal:", obj1.detect_loop())

print("\n--> Linked List After Removing Loop:")
obj1.display()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None

class DoublyLinkedList:
    def __init__(self):
        self.head=None
    
    def insert_at_beginning(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head.prev=new_node
            self.head=new_node
    
    def insert_at_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.next is not None:
                temp=temp.next
            temp.next=new_node
            new_node.prev=temp
    
    def delete_at_position(self,position):
        if self.head is None:
            print("List is empty.")
            return
        temp=self.head
        if position==1:
            self.head=temp.next
            if self.head:
                self.head.prev=None
            temp=None
            return
        for _ in range(position-1):
            if temp is None:
                print("Position out of bounds.")
                return
            temp=temp.next
        if temp is None:
            print("Position out of bounds.")
            return
        if temp.next:
            temp.next.prev=temp.prev
        if temp.prev:
            temp.prev.next=temp.next
        temp=None
    
    def traverse_forward(self):
        temp=self.head
        if temp is None:
            print("List is empty.")
            return
        while temp is not None:
            print(temp.data,end=" <-- --> ")
            temp=temp.next
        print("None")
    
    def traverse_reverse(self):
        temp=self.head
        if temp is None:
            print("List is empty.")
            return
        while temp.next is not None:
            temp=temp.next
        while temp is not None:
            print(temp.data,end=" <-- --> ")
            temp=temp.prev
        print("None")

obj1 = DoublyLinkedList()
obj1.insert_at_beginning(10)
obj1.insert_at_beginning(20)
obj1.insert_at_end(30)
obj1.insert_at_end(40)

print("Forward Traversal:")
obj1.traverse_forward()

print("\nReverse Traversal:")
obj1.traverse_reverse()

obj1.delete_at_position(2)
print("\nAfter Deleting Node at Position 2:")
obj1.traverse_forward()
```
