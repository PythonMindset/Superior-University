```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None

class Linked_list:
    def __init__(self):
        self.head=None

    def add_last(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.ref is not None:
                temp=temp.ref
            temp.ref=new_node
        return new_node
    
    def create_loop(self, position):
        if position<1:
            return
        loop_start=None
        temp=self.head
        count=1
        while temp.ref is not None:
            if count==position:
                loop_start = temp
            temp=temp.ref
            count+=1
        if loop_start:
            temp.ref=loop_start
    
    def detect_loop(self):
        slow = self.head
        fast = self.head
        while fast is not None and fast.ref is not None:
            slow=slow.ref
            fast=fast.ref.ref
            if slow==fast:
                return True
        return False
    
    def find_loop_start(self):
        slow=self.head
        fast=self.head

        while fast and fast.ref:
            slow=slow.ref
            fast=fast.ref.ref

            if slow==fast:
                slow=self.head
                while slow!=fast:
                    slow=slow.ref
                    fast=fast.ref
                return slow
        return None
    
    def remove_loop(self):
        loop_start=self.find_loop_start()
        if loop_start is None:
            return
        temp=loop_start
        while temp.ref!=loop_start:
            temp=temp.ref
        temp.ref=None
    
    def display(self, limit=15):
        visited=set()
        temp=self.head
        count=0
        while temp and count<limit:
            if temp in visited:
                print(f"{temp.data} (loop starts here)")
                break
            print(temp.data,end=" -> ")
            visited.add(temp)
            temp=temp.ref
            count+=1
        else:
            print("None")
    
obj1 = Linked_list()
obj1.add_last(10)
obj1.add_last(20)
node30 = obj1.add_last(30)
obj1.add_last(40)
obj1.add_last(50)

obj1.create_loop(3)

print("--> Linked List Before Removing Loop:")
obj1.display()

print("\nLoop detected:", obj1.detect_loop())

loop_node = obj1.find_loop_start()
if loop_node:
    print("Loop starts at node:", loop_node.data)

obj1.remove_loop()

print("\nLoop detected after removal:", obj1.detect_loop())

print("\n--> Linked List After Removing Loop:")
obj1.display()
```
