```python
import time

class linear_queue:
    def __init__(self):
        self.queue=[]
        self.limit=5
    
    def enqueue(self,element):
        if self.limit <= len(self.queue):
            print("Queue is full.")
        else:
            self.queue.append(element)
    
    def dequeue(self):
        self.queue.pop(0)
    
    def front(self):
        return self.queue[0]
    
    def rear(self):
        return self.queue[-1]
    
    def is_empty(self):
        if len(self.queue) == 0:
            return True
        else:
            return False
    
    def is_full(self):
        if self.limit == len(self.queue):
            return True
        else:
            return False

class cirular_queue:
    def __init__(self,limit=5):
        self.queue=[None]*limit
        self.limit=limit
        self.front=-1
        self.rear=-1
    
    def enqueue(self,element):
        if (self.rear+1)%self.limit==self.front:
            print("Queue is full.")
        else:
            if self.front== -1:
                self.front=0
            self.rear=(self.rear+1)%self.limit
            self.queue[self.rear]=element
    
    def dequeue(self):
        if self.front== -1:
            print("Queue is empty.")
        else:
            dequeued_element=self.queue[self.front]
            self.queue[self.front]=None
            if self.front==self.rear:
                self.front= -1
                self.rear= -1
            else:
                self.front=(self.front+1)%self.limit
            return dequeued_element
    
    def front_element(self):
        if self.front== -1:
            print("Queue is empty.")
            return None
        return self.queue[self.front]
    
    def rear_element(self):
        if self.rear== -1:
            print("Queue is empty.")
            return None
        return self.queue[self.rear]
    
    def is_empty(self):
        return self.front== -1
    
    def is_full(self):
        return (self.rear+1)%self.limit==self.front

def comparison():
    print("For Python List")
    start = time.time()
    obj1=linear_queue()
    obj1.enqueue(1)
    obj1.enqueue(2)
    obj1.enqueue(3)
    obj1.front()
    obj1.rear()
    obj1.is_empty()
    obj1.is_full()
    end = time.time()
    print(f"Time Taken: {end-start: .10f} seconds.")

    print("For Linked List")
    start = time.time()
    obj2=cirular_queue()
    obj2.enqueue(1)
    obj2.enqueue(2)
    obj2.enqueue(3)
    obj2.dequeue()
    obj2.front_element()
    obj2.rear_element()
    obj2.is_empty()
    obj2.is_full()
    end = time.time()
    print(f"Time Taken: {end-start: .10f} seconds.")
comparison()
```
