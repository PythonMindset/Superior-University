### Task 2

```python
def fibonacci_recursive(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    
    series = fibonacci_recursive(n - 1)
    series.append(series[-1] + series[-2])
    return series

def fibonacci_iterative(n):
    a, b = 0, 1
    for i in range(n):
        print(a,end=" ")
        a, b = b, a + b

def comparison():
    n = 20
    start = time.time()
    print(fibonacci_recursive(n))
    end = time.time()
    print(f"Recursive Fibonacci Time: {end-start: .10f} seconds.")

    start = time.time()
    fibonacci_iterative(n)
    end = time.time()
    print(f"\nIterative Fibonacci Time: {end-start: .10f} seconds.")
comparison()
```

### Task 3

```python
import matplotlib.pyplot as plt
import numpy as np

# Input sizes
n = np.arange(1, 1001)

# Complexity functions
def constant(n):
    return np.ones_like(n)

def log_n(n):
    return np.log2(n)

def linear(n):
    return n

def n_log_n(n):
    return n * np.log2(n)

def quadratic(n):
    return n ** 2

def exponential(n):
    return 2 ** (n / 100)  # Scaled down for visualization

# Compute values
complexities = {
    'O(1)': constant(n),
    'O(log n)': log_n(n),
    'O(n)': linear(n),
    'O(n log n)': n_log_n(n),
    'O(n^2)': quadratic(n),
    'O(2^n)': exponential(n)
}

# Plot
plt.figure(figsize=(12, 8))
for label, values in complexities.items():
    plt.plot(n, values, label=label)

plt.title("Big-O Time Complexity Visualization")
plt.xlabel("Input Size (n)")
plt.ylabel("Operations")
plt.legend()
plt.grid(True)
plt.ylim(0, 1000)  # limit y-axis for better clarity
plt.tight_layout()
plt.show()
```
